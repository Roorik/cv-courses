```python
# импорт либ
import logging
import numpy as np
import matplotlib.pyplot as plt

# настраиваем логгирование
logging.basicConfig(filename='log.txt', level=logging.INFO)
log = logging.getLogger(__name__)

def drow_graph(x: np.ndarray, y1: np.ndarray, y2: np.ndarray) -> None:
    
    '''
    Рисует 2 графика и закрашивает заданное пространство
    
    :param x: np.ndarray - массив точек
    :param y1: np.ndarray - первая функция синусоида 
    :param y2: np.ndarray - вторая функция синусоида 
    '''
       
    try:
        # создаём графики
        fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))

        # *первый график
        # первая синусоида
        ax1.plot(x, y1, color='b', label='sin(x)')
        ax1.set_title('Первый график')
        ax1.grid(visible=True, linestyle='--', alpha=0.5)
        ax1.legend()

        # вторая синусоида
        ax1.plot(x, y2, color='r', label='sin(0.4 * np.pi * x)')
        ax1.set_title('Синусоида')
        ax1.grid(visible=True, linestyle='--', alpha=0.5)
        ax1.legend()
        
        # закрашиваем пространство между двумя графиками
        ax1.fill_between(x, y1, y2, color='m') 
       
        # *второй график
        ax2.plot(x, y1, color='g', label='sin(x)')
        ax2.set_title('Второй график')
        ax2.grid(visible=True, linestyle='--', alpha=0.5)
        ax2.legend()
        
        # закрашиваем пространство под графиком
        ax2.fill_between(x, y1, -1, color='y') 
        
        # общая настройка
        fig.suptitle('Примеры графиков', fontsize=14)               
        
        plt.tight_layout()  
        plt.show()
    except Exception as e:
        print('Ошибка')
        log.error(f'Произошла ошибка: {e}')
    
def main():
    # объявляем:
    x = np.linspace(0, 10, 100) # массив ста равноразнесённых точек от 0 до 10
    y1 = np.sin(x) # первая синусоида
    y2 = np.sin(0.4 * np.pi * x) # вторая синусоида
    
    # отрисовываем графики функций
    drow_graph(x=x, y1=y1,  y2=y2)
    
if __name__ == '__main__':
    main()
        
```
