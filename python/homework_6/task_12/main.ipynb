```python
# импорт либ
import logging
import numpy as np
import matplotlib.pyplot as plt

# настраиваем логгирование
logging.basicConfig(filename='log.txt', level=logging.INFO)
log = logging.getLogger(__name__)

def drow_graph(x: np.ndarray, y1: np.ndarray, y2: np.ndarray) -> None:
    
    '''
    Рисует график c второстепенной осью Y
    
    :param x: np.ndarray - массив точек
    :param y1: np.ndarray - функция синусоида 
    :param y2: np.ndarray - функция квадратичной функции
    '''
       
    try:
        # создаём график
        fig, ax1 = plt.subplots(figsize=(10, 6))

        # график синусоида
        color = 'm'
        
        ax1.plot(x, y1, color=color)
        ax1.set_ylabel('sin(x)', color=color)
        ax1.tick_params(axis='y', labelcolor=color)
        
        # создаём новую ось скопировав первую
        ax2 = ax1.twinx()
        
        # квадратичный график
        color = 'b'
        
        ax2.plot(x, y2, color=color)
        ax2.set_ylabel('x^2', color=color)
        ax2.tick_params(axis='y', labelcolor=color)

        # общая настройка
        fig.suptitle('Примеры графиков', fontsize=14)
        
        plt.tight_layout()  
        plt.show()
    except Exception as e:
        print('Ошибка')
        log.error(f'Произошла ошибка: {e}')
    
def main():
    # объявляем:
    x = np.linspace(0, 10, 100) # массив ста равноразнесённых точек от 0 до 10
    y1 = np.sin(np.pi * x) # синусоида
    y2 = x**2 # квадратичная функция

    # отрисовываем графики функций
    drow_graph(x=x, y1=y1, y2=y2)
    
if __name__ == '__main__':
    main()
```
