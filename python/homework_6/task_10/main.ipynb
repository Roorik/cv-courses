```python
# импорт либ
import logging
import numpy as np
import matplotlib.pyplot as plt

# настраиваем логгирование
logging.basicConfig(filename='log.txt', level=logging.INFO)
log = logging.getLogger(__name__)

def drow_graph(x: np.ndarray, y1: np.ndarray, y2: np.ndarray, y3: np.ndarray, y4: np.ndarray) -> None:
    
    '''
    Рисует 4 графика функций
    
    :param x: np.ndarray - массив точек
    :param y1: np.ndarray - функция синусоида 
    :param y2: np.ndarray - функция косиноида
    :param y3: np.ndarray - функция квадратичной функции
    :param y4: np.ndarray - функция квадратного корня
    '''
       
    try:
        # создаём график
        fig, ax = plt.subplots(2, 2, figsize=(10, 6))

        # график синусоида
        ax[0, 0].plot(x, y1, color='m', label='sin(x)')
        ax[0, 0].set_title('Синусоида')
        ax[0, 0].grid(visible=True, linestyle='--', alpha=0.5)
        ax[0, 0].legend()

        # график косинусоида
        ax[0, 1].plot(x, y2, color='g', label='cos(x)')
        ax[0, 1].set_title('Косинусоида')
        ax[0, 1].grid(visible=True, linestyle='--', alpha=0.5)
        ax[0, 1].legend()

        # график квадратичной функции
        ax[1, 0].plot(x, y3, color='r', label='x^2')
        ax[1, 0].set_title('Квадратичная функция')
        ax[1, 0].grid(visible=True, linestyle='--', alpha=0.5)
        ax[1, 0].legend()

        # график квадратного корня
        ax[1, 1].plot(x, y4, color='b', label='sqrt(x)')
        ax[1, 1].set_title('Квадратный корень')
        ax[1, 1].grid(visible=True, linestyle='--', alpha=0.5)
        ax[1, 1].legend()

        # общая настройка
        fig.suptitle('Примеры графиков', fontsize=14)
        
        plt.tight_layout()  
        plt.show()
    except Exception as e:
        print('Ошибка')
        log.error(f'Произошла ошибка: {e}')
    
def main():
    # объявляем:
    x = np.linspace(0, 10, 100) # массив ста равноразнесённых точек от 0 до 10
    y1 = np.sin(x) # синусоида
    y2 = np.cos(x) # косинусоида
    y3 = x**2 # квадратичная функция
    y4 = np.sqrt(x) # квадратный корень
    
    # отрисовываем графики функций
    drow_graph(x=x, y1=y1, y2=y2, y3=y3, y4=y4)
    
if __name__ == '__main__':
    main()
```
